```python
import matplotlib.pyplot as plt 
```

# Parallel Overhead 

```python
W_options = [200, 400, 800]
n_options = [2, 9, 14]
```

```python
n_and_speedups = [
(2, [1.189064007719524, 1.3353836921461257, 0.9695895267966727]),
(9, [1.3511145142079293, 1.0786045102367658, 0.9546867010187626]),
(14, [1.2342071494042164, 0.9894206549118388, 0.925192912260646])]


for (n, speedup) in n_and_speedups:
    plt.plot(W_options,
    [1 / elem for elem in speedup],
    marker='x'
    )

plt.legend(["n=2", "n=9", "n=14"])
plt.title("Parallel Overhead Experiment")
plt.xlabel("Work per Packet (W)")
plt.ylabel("Ratio Serial-Queue to Serial")
```
